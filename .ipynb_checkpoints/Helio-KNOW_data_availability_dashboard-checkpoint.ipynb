```python
import os
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)

import datetime

import csv

import json

import plotly.express as px

```

```python
load_directory = '/Users/ryanmcgranaghan/Documents/Helio_ECIP/data/summary_data/'

metadata_file_helioKNOW = '/Users/ryanmcgranaghan/Documents/Helio_ECIP/dev/Helio-KNOW/helioKNOW_metadata.json'
```

```python
with open(metadata_file_helioKNOW) as json_file:

    metadata_helioKNOW = json.load(json_file)
    
metadata_helioKNOW
```

##### Select date range 
(future: this could be selected by the user from a calendar widget)

```python
date_range_availability = [datetime.datetime(2000,1,1),
                           datetime.datetime(2000,12,31)]

```

```python
data_avail = pd.DataFrame(index=pd.date_range(date_range_availability[0],date_range_availability[1],freq='D').tolist(),
                          columns=metadata_helioKNOW['platforms'])
data_avail.index.name='date'
data_avail
```

```python
num_possible_data_points = 24*60. # possible number of observations in the daily files (1440 minutes)
for d in data_avail.index:
    print('working on date = {}'.format(d))
    yr_tmp = d.year
    mo_tmp = d.month
    dy_tmp = d.day
    for p in data_avail.columns.to_list():
        print('    working on platform = {}'.format(p))
        
        if 'dmsp' in p:
            file_tmp = os.path.join(load_directory,
                                    str(yr_tmp),
                                    str(yr_tmp)+str(mo_tmp).zfill(2)+str(dy_tmp).zfill(2),
                                    p,
                                    p[0:4]+'_'+p[4:9]+'_ephemeris_'+str(yr_tmp)+str(mo_tmp).zfill(2)+str(dy_tmp).zfill(2)+'.csv')
        elif 'fast' in p:
            file_tmp = os.path.join(load_directory,
                                    str(yr_tmp),
                                    str(yr_tmp)+str(mo_tmp).zfill(2)+str(dy_tmp).zfill(2),
                                    p,
                                    p[0:4]+'_ephemeris_'+str(yr_tmp)+str(mo_tmp).zfill(2)+str(dy_tmp).zfill(2)+'.csv')
        else:
            print('platform not recognized...skipping')
            continue
        
        print(file_tmp)
        if os.path.isfile(file_tmp):
            with open(file_tmp) as fp:
                count = 0
                for _ in fp:
                    count += 1
            
            data_avail.loc[d,p] = np.floor( (count / num_possible_data_points) * 100 )
            
```

```python
data_avail = data_avail.fillna(0)
```

#### Visualize the availability

```python
data_avail_melt = data_avail.copy()
data_avail_melt=pd.melt(data_avail_melt.reset_index(), id_vars=['date'], value_vars=data_avail_melt.columns)
data_avail_melt
```

```python
# plotly express
#  Code benefitted from basis provided by: https://stackoverflow.com/questions/59602653/data-availability-chart-in-python


fig = px.bar(data_avail_melt, 
             x="date", 
             y="variable", 
             color='value', orientation='h',
             hover_data=["date"],
             height=600,
             color_continuous_scale=['firebrick', '#2ca02c'],
             title='Data Availabiltiy Plot',
             template='plotly_white',
            )


fig.update_layout(
    yaxis = dict(
        title='Platform'
    ),
    xaxis = dict(
        title='Date',
    ),
    coloraxis= dict(
        colorbar_title='Percentage Available'
    ),
)

fig.show()
```
