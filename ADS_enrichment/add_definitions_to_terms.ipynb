```python
# Experimenting with adding definitions to UAT
        
       
from rdflib import ConjunctiveGraph, Namespace, URIRef
from collections import deque
import pandas as pd

# === Load TriG dataset as a ConjunctiveGraph (handles named graphs) ===
file_path = "/Users/ryanmc/Downloads/pp_project_aasunifiedastronomythesaurustest.trig"  # adjust to your path
g = ConjunctiveGraph()
g.parse(file_path, format="trig")

# === Define SKOS namespace ===
SKOS = Namespace("http://www.w3.org/2004/02/skos/core#")

# === Exact Heliophysics URI ===
HELIOS = URIRef("http://astrothesaurus.org/AASUnifiedAstronomyThesaurusTEST/2374")

# --- Helper: get prefLabel ---
def get_label(u):
    labels = list(g.triples((u, SKOS.prefLabel, None)))
    if not labels:
        return None
    # Prefer English if available
    for _, _, lbl in labels:
        if getattr(lbl, "language", None) == "en":
            return str(lbl)
    return str(labels[0][2])

# === Traverse hierarchy ===
visited = set()
results = []
queue = deque()

# Seed traversal:
for child, _, _ in g.triples((None, SKOS.broader, HELIOS)):
    queue.append(child)
for _, _, child in g.triples((HELIOS, SKOS.narrower, None)):
    queue.append(child)
for _, _, child in g.triples((HELIOS, SKOS.hasTopConcept, None)):
    queue.append(child)

# Breadth-first search
while queue:
    node = queue.popleft()
    if node in visited:
        continue
    visited.add(node)

    label = get_label(node)
    results.append((label, str(node)))

    # Expand children of this node
    for child, _, _ in g.triples((None, SKOS.broader, node)):
        if child not in visited:
            queue.append(child)
    for _, _, child in g.triples((node, SKOS.narrower, None)):
        if child not in visited:
            queue.append(child)

# === Sort results nicely ===
results_sorted = sorted(results, key=lambda x: (x[0] is None, (x[0] or "").lower()))

# # === Save to CSV ===
df = pd.DataFrame(results_sorted, columns=["Label", "URI"])
# df.to_csv("heliophysics_terms.csv", index=False, encoding="utf-8")

print(f"Extracted {len(df)} terms under Heliophysics")
print("Example:", df.head())

```
